```python
import pandas as pd
```

```python
# 36-38라운드에 대한 승/무승 예측
testdata = pd.read_csv('C:/Users/dh65a/Downloads/3638round.csv', sep=',', header = 0, engine = 'python')
```

```python
testdata
```

```python
#패(L)와 무(D)를 무승(NW)로 변경. sf 칼럼 삭제
testdata = testdata.replace({'result':"L"}, 'NW')
testdata = testdata.replace({'predict':"L"}, 'NW')
testdata = testdata.replace({'result':'D'}, 'NW')
testdata = testdata.replace({'predict':'D'}, 'NW')
testdata = testdata.drop(columns = ['sf'], axis = 1)
```

```python
df = pd.DataFrame(testdata)
df
```

```python
#승, 무승 예측 성공률
print(round(len(df.loc[df['predict'] == df['result']])/len(df)*100, 2), '%')
```
