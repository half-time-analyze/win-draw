```python
import pandas as pd
```

```python
data1 = pd.read_csv('C:/half_time_analyze/win-draw/final/soccer.csv', sep = ',', header = 0, engine = 'python')
data2 = pd.read_csv('C:/half_time_analyze/win-draw/final/soccer2.csv', sep = ',', header = 0, engine = 'python')
data3 = pd.read_csv('C:/half_time_analyze/win-draw/final/soccer3.csv', sep = ',', header = 0, engine = 'python')
```

```python
# csv 파일 병합
total_data = pd.concat([data1, data2, data3])
total_data
```

```python
total_data.columns = total_data.columns.str.replace(' ', '_')
total_data.columns
```

```python
# 전반전 슈팅값 변수 추가
total_data['Shots_1st_Half'] = total_data['Shots_Off_1st_Half'] + total_data['Shots_On_1st_Half']
```

```python
# homeAway에서 home은 1, Away는 0으로 설정
total_data = total_data.replace({'homeAway': 2}, 0)
total_data
```

```python
# 인덱스 재설정
total_data = total_data.reset_index(drop=True)
total_data
```

```python
# 독립변수와 종속변수 추출
total_data = total_data[['homeAway', 'Ball_Possession_1st_Half', 'Corners_1st_Half', 'Fouls_Commited_1st_Half', 'Goals_1st_Half', 'Red_Cards_1st_Half', 'Shots_1st_Half', 'Shots_On_1st_Half', 'Yellow_Cards_1st_Half', 'Result']]
total_data
```

```python
#무, 패 데이터 추출
idx = total_data[total_data['Result'] == "L"].index
total_data.drop(idx, inplace = True)
total_data
```

```python
# 승, 무승부는 1로, 패배는 0으로 값 변경
total_data = total_data.replace({'Result': 'W'}, 1)
total_data = total_data.replace({'Result': 'D'}, 0)
total_data
```

```python
#결측치 확인
total_data.info()
```

```python
#데이터 내보내기
total_data.to_csv('soccer_data.csv', index = False)
```

```python
#타겟 데이터 따로 저장
target = total_data['Result']
#total_data에서 종속 변수 제거
total_data = total_data.drop(columns=['Result'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# X, Y 설정
Y = target
# X = soccer_half_scaled
X = total_data
```

```python
# 훈련 데이터와 테스트 데이터 분리
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 360, stratify = Y)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
print(round(sum(Y_train)/len(Y_train)*100, 2), '%')
```

```python
print(round(sum(Y_test)/len(Y_test)*100, 2), '%')
```

```python
import statsmodels.api as sm
```

```python
#절편(intercept) 추가
X_ = sm.add_constant(X_train, has_constant='add')
```

```python
logit_m = sm.Logit(Y_train, X_)
logit_m_train = logit_m.fit()
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = logit_m_train.predict(sm.add_constant(X_test)) > 0.5
accuracy = accuracy_score(Y_test, y_pred)
print('정확도: {0:.3f}'.format(accuracy))
```

```python
logit_m_train.summary()
```

- coef : 계수
